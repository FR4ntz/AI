```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets
```

LOAD DATASET

```python
digits = datasets.load_digits()
print("Jumlah data:", len(digits.data))
print ("Jumlah fitur per data:", len(digits.data[0]))
```

TAMPILKAN GAMBAR

```python
plt.imshow(digits.images[1], cmap='gray')
plt.title("Contoh Gambar Angka")
plt.show()
```

LATIH MODEL SEDERHANA (TANPA TEORI)

```python
from sklearn.linear_model import LogisticRegression
x = digits.data
y = digits.target
model = LogisticRegression(max_iter=1000)
model.fit(x,y)
print("Model Berhasil dilatih!")
```

MINTA MODEL MENEBAK

```python
prediksi = model.predict(x[:10])
print("Prediksi:", prediksi)
print("Jawaban asli:", y[:10])
```
